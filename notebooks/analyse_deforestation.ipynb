## 🛰️ 1. Analyse du Premier Groupe de Parcelles (Lot 1)
Nous allons demander à l'IA de comparer les images 'avant' (Lot A) et 'après' (Lot B) pour le premier secteur.

```python
# On lance l'analyse automatisée sur toutes les images du Lot 1
df1 = pipe.run_full_forest_analysis('../data', 'Lo1_A', 'Lot1_B', 'lot1')
df1.head()
```

### 📊 Résultats détaillés pour le Lot 1
Le tableau ci-dessous montre la surface de forêt mesurée au début (t0) et à la fin (t1). 
- Une **Variation négative** indique une perte de forêt (déforestation).
- Le **Diagnostic** est le verdict rendu par l'IA selon la gravité de la situation.

```python
# Mise en forme pour rendre le tableau plus lisible
# Note : on utilise .rename() pour éviter les erreurs de décalage de colonnes
df1_styled = df1.rename(columns={
    'image': 'Zone',
    'surface_t0': 'Surface Début',
    'surface_t1': 'Surface Fin',
    'loss_percentage': 'Variation %',
    'Verdict': 'Verdict Expert'
})

# On applique un dégradé de couleur uniquement sur la colonne de variation numérique
df1_styled.style.background_gradient(cmap='RdYlGn_r', subset=['Variation %'])\
        .format({'Variation %': '{:.2f}%'})
```
