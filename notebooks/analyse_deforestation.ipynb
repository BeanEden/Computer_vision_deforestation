### 🏁 Conclusion de l'Analyse
Les résultats ci-dessus sont également exportés automatiquement dans le dossier `outputs/` sous forme de fichiers CSV et de cartes de changement. 

Ces données alimentent directement le **Dashboard interactif** pour une visualisation simplifiée.

Méthodes Utilisées
1. Prétraitement & Signal
>- Filtrage Anti-Bruit (Gaussien) : Réduction du bruit thermique et impulsionnel des capteurs satellites. Justifié par une analyse du PSNR (Peak Signal-to-Noise Ratio).
>- Normalisation Radiométrique (CLAHE) : Limitation du contraste adaptatif dans l'espace de couleur Lab pour compenser les variations d'albédo dues à l'heure d'acquisition.
2. Extraction de Primitives (Features)
>- Multispectralité : Utilisation des canaux RGB et HSV pour isoler la signature de la chlorophylle.
>- Indices Végétaux : Calcul du Ratio Vert (Green/Total) pour accentuer la biomasse.
>- Texture de Second Ordre (GLCM) : Analyse de la matrice de cooccurrence pour évaluer le contraste et la corrélation locale des massifs.
3. Segmentation par IA (Clustering)
>- K-Means (Non-supervisé) : Classification automatique des pixels en 3 strates (Canopée, Sol Nu, Ombrage).
>- CAH (Clustering Ascendant Hiérarchique) : Utilisation de dendrogrammes sur échantillons pour valider mathématiquement le nombre optimal de clusters (k).
>- Score de Silhouette : Mesure de la cohésion des segments pour garantir une séparation nette entre forêt et zones dégradées.
4. Morphologie Mathématique
>- Ouverture (Opening) : Élimination des pixels isolés ("bruit de segmentation").
>- Fermeture (Closing) : Consolidation des houppiers pour assurer la continuité spatiale du massif forestier.
>- Élément Structurant : Noyau carré 5x5 pour respecter les lisières naturelles.

⚠️ Problématiques Adressées

>- Bruit Capteur : Application sélective du filtre Gaussien (optimisé via PSNR).
>- Variabilité Lumineuse : Passage en espace Lab pour traiter la luminance séparément.
>- Confusion Sol/Canopée : Hybridation des features (Couleur + Texture GLCM).
>- Fragmentation des données : Post-traitement morphologique pour "lisser" les massifs.
>- Passage à l'échelle : Architecture modulaire (Lot 1, Lot 2) avec isolation des outputs.
>- Communication Expert : Refonte lexique (Sylviculture, Canopée vs Tech-speak).
